### Предварительная очистка текста от слов и лишних символов
Из-за того, что некоторые документы содержат данные, представленные в табличном виде, из-за чего парсинг выходит не очень корректным, то необходимо удалить (заменить на пробел) такие символы как \n, \r, \t и их комбинации, удалить слова 'экзамен', 'зачёт' и другие.

```python
regs_2 = [
    r'Тема \d\d?[.]',
    r' {2,}',
]

def process_text(text, regs):
    text = text.encode("ansi", "ignore").decode('ansi')
    text_ = text.replace('\n', '').replace('\t', '').replace('ё', 'е')
    for reg in regs:
        text_ = re.sub(reg, ' ', text_)
    text_ = re.sub(r' {2,}', ' ', text_)
    return text_
```

```python
new_text = process_text(text, regs_2)
new_text
```

### Выделение дисциплин из текста

```python
def getCourses(text, split_by, alert=False):
    course_dict = {}
    courses = re.findall(split_by, text)
    splitted = re.split(split_by, text)[1:]
    for i in range(len(courses)):
        c = courses[i][10:][:-3]
        c = re.sub('Б1[.][ОВ][.]\d\d[.]?', '', c)
        c = re.sub('Б1[.]В[.]ДВ[.]\d\d[.]\d\d[.]?', '', c)
        c = re.sub('ФТД[.]В[.]\d\d[.]?', '', c).replace('«', '').replace('»', '')
        c = re.sub('\(.*?\)', '', c)
        c = c.lower().lstrip().rstrip()
        course_dict[c] = splitted[i][2:]

    return course_dict
    
course_dict = getCourses(new_text, r'АННОТАЦИЯ.*?«?.*?»? Цел', alert=True)
courses = list(course_dict.keys())
```

```python
ind = file.rfind('\\')

for k in course_dict:
    path = file[:ind - 3] + '\\' + k + '.txt'
    
    file_ = open(path, 'w+', encoding='ansi')
    file_.write(course_dict[k])
    file_.close()
```
